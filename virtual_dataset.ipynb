```python
# nomi 
names: list = ['name_'+str(i+1) for i in range (n)]
df = pd.DataFrame(index = names)
df = df.sample(frac = 1) #shuffles the dataframe

#genere
df['rnd'] = np.random.randint(100, size = n)
df['gender'] = np.where(df['rnd']<=47, 'm', 'f')
df.drop(columns=['rnd'], inplace=True)

```

```python
# destinazioni
df['rnd'] = np.random.randint(100, size = n) # valori casuali 
df['dest'] = np.where(df['rnd']<=41, 'P', 'B')
df.drop(columns=['rnd'], inplace=True)

# origini
df['orig'] = np.random.randint(24, size = n) # valori casuali 

df.sample(frac = 0.004)
```

```python
def assign_values(df, category, values_dict:dict):
    sumvalues = 0 
    start = 0
    for key,value in values_dict.items():
        print('assigning ' , key, value)
        end = start + int(n*value)
        df[category][start: end] = key
        start = end
        sumvalues += value
    assert sumvalues > 0.9999999, print(sumvalues)
    assert (df[category] == '-').mean() == 0.0
    df = df.sample(frac = 1) #shuffles the dataframe
    return df
```

```python
# propensione al cambiamento
prop_bici:dict = {'0':0.80, '1':0.10,'2':0.04, '3':0.03,'4':0.02, '5':0.01 }
df['pro_bici'] = '-'
df = assign_values(df, 'pro_bici', prop_bici)

prop_bus:dict = {'0':0.60, '1':0.10,'2':0.08, '3':0.06,'4':0.05, '5':0.11 }
df['pro_bus'] = '-'
df = assign_values(df, 'pro_bus', prop_bus)

prop_sw:dict = {'0':0.11, '1':0.05,'2':0.04, '3':0.20,'4':0.40, '5':0.20 }
df['pro_sw'] = '-'
df = assign_values(df, 'pro_sw', prop_sw)

df.sample(frac = 0.01)
```
